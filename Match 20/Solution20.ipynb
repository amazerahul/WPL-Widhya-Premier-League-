### Overall Player Performance

```python
def opp_batsman(df):
    total_runs_scored = df.groupby(['match_id', 'batsman'])['batsman_runs'].sum()
    balls_faced = df.groupby(['match_id', 'batsman'])['ball'].count()
    boundary_runs = df[df['batsman_runs'].isin([4, 5, 6, 7])].groupby(['match_id', 'batsman'])['batsman_runs'].sum()
    non_boundary_runs = df[df['batsman_runs'].isin([1, 2, 3])].groupby(['match_id', 'batsman'])['batsman_runs'].sum()
    opp = (((total_runs_scored / balls_faced) + boundary_runs) - (non_boundary_runs / balls_faced))
    opp_df = pd.DataFrame(opp)
    opp_df.fillna(0, inplace=True)
    min_opp = min(opp)
    max_opp = max(opp)
    normalized_opp = (opp-min_opp)/(max_opp - min_opp) 
    return normalized_opp
```

```python
def opp_bowler(df):
    boundary_runs = df[df['total_runs'].isin([4, 5, 6, 7])].groupby(['match_id', 'bowler'])['total_runs'].sum()
    boundary_hits = df[df['total_runs'].isin([4, 5, 6, 7])].groupby(['match_id', 'bowler'])['ball'].count()
    balls_bowled = df.groupby(['match_id', 'bowler'])['ball'].count()
    opp = ((boundary_runs / boundary_hits) / balls_bowled)
    min_opp = min(opp)
    max_opp = max(opp)
    normalized_opp = (opp-min_opp)/(max_opp - min_opp) 
    return normalized_opp
```

```python
opp_batsman = pd.DataFrame(opp_batsman(sub_data[sub_data['season']>2015])).reset_index()
```

```python
opp_bowler = pd.DataFrame(opp_bowler(sub_data[sub_data['season']>2015])).reset_index()
```

```python
opp_player = pd.merge(opp_batsman, opp_bowler, left_on=['match_id', 'batsman'], right_on=['match_id', 'bowler'], how='outer')
```

```python
opp_player['batsman'].fillna('NA', inplace=True)
opp_player['bowler'].fillna('NA', inplace=True)
```

```python
opp_player['player'] = np.where(opp_player['batsman']=='NA', opp_player['bowler'], opp_player['batsman'])
```

```python
opp_player.drop(['batsman', 'bowler'], axis=1, inplace=True)
```

```python
cols = ['match_id', 'player', '0_x', '0_y']
opp_player = opp_player[cols]
```

```python
opp_player.rename(columns = {"0_x" : "batting_opp", "0_y" : "bowling_opp"}, inplace=True)
```

```python
opp_player[['batting_opp', 'bowling_opp']] = opp_player[['batting_opp', 'bowling_opp']].fillna(0)
```

```python
opp_player['total_opp'] = opp_player['batting_opp'] + opp_player['bowling_opp']
```

```python
opp_player = opp_player.sort_values(by='total_opp', ascending=False)
```

```python
opp_player.set_index(['match_id', 'player'], inplace=True)
```

### Balls M Shami will take to get his first wicket??

```python
def first_wicket(df):
    balls = 1
    
    for i in df.player_dismissed_bool:
        if i == 0:
            balls += 1
        else:
            break
    return balls
```
