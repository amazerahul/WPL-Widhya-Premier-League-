### How many balls SRH take to reach 50 runs on team total?

```python
def Check(df):
    balls = 0
    runs = 0
    for i in df.total_runs:
        if runs >= 50:
            break
        else:
            runs += i
            balls += 1
    return balls
```

```python
sub_data5 = data[data['batting_team']=='Sunrisers Hyderabad']
```

```python
sub_data5.groupby(['match_id']).apply(Check).plot.bar(title='Balls taken by SRH to score 50 runs', figsize=(20,5))
```

```python
sub_data5.groupby(['match_id']).apply(Check).plot.box(title='Boxplot distribution of balls taken by SRH to score 50 runs')
```

```python
sub_data5.groupby(['match_id']).apply(Check).describe()
```

```python
sub_data[sub_data['batting_team']=='Sunrisers Hyderabad'].groupby('match_id').apply(Check).plot.bar(title='Balls taken by SRH to score 50 runs against KXIP')
```

```python
sub_data[sub_data['batting_team']=='Sunrisers Hyderabad'].groupby('match_id').apply(Check).plot.box(title='Boxplot distribution of balls taken by SRH to score 50 runs against KXIP')
```

```python
sub_data[sub_data['batting_team']=='Sunrisers Hyderabad'].groupby('match_id').apply(Check).describe()
```

```python
sub_data[(sub_data['batting_team']=='Sunrisers Hyderabad') & (sub_data['match_id'] == 11339)].groupby('batsman')['batsman_runs'].sum().plot.bar()
```

### How many runs will Chris Gayle score in the first 10 balls he faces?

```python
def Check1(df):
    balls = 0
    runs = 0
    for i in df.batsman_runs:
        if balls == 10:
            break
        else:
            runs += i
            balls += 1
    return runs
```

```python
sub_data6 = data[data['batsman']=='CH Gayle']
```

```python
sub_data6.groupby(['match_id']).apply(Check1).plot.bar(title='Runs scored by CH Gayle in first 10 balls', figsize=(15,5))
```

```python
dummy = pd.DataFrame(sub_data6.sort_values(['match_id','over', 'ball']).groupby('match_id').head(10).groupby('match_id')['ball'].count())
```

```python
balls10_index = dummy[dummy['ball']==10].index
```

```python
less10_index = dummy[dummy['ball']<10].index
```

```python
sub_data6[(sub_data6['match_id'].isin(balls10_index)) & (sub_data6['season']>2015)].groupby(['match_id']).head(10).groupby('match_id')['batsman_runs'].sum().plot.box(title='Runs scored by CH Gayle when he played 10 balls')
```

```python
sub_data6[(sub_data6['match_id'].isin(less10_index)) & (sub_data6['season']>2015)].groupby(['match_id']).head(10).groupby('match_id')['batsman_runs'].sum().plot.box(title='Runs scored by CH Gayle when he played less than 10 balls')
```
