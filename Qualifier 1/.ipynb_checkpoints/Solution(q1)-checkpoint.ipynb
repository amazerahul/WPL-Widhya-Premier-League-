### Economical Bowlers(< 9 economy)

```python
def Repeat(df, filters):
    for i, j in filters:
        df = df.drop(df[(df['match_id']==i) & (df['bowler']==j)].index)
    return df
```

```python
def Economy_bowlers(df):
    ### Less than or equal to 1 over over bowled filter
    overs_bowled = pd.DataFrame(df.groupby(['match_id', 'bowler'])['over'].unique().apply(lambda x : len(x)))
    over_bowled_index = overs_bowled[overs_bowled['over']<=1].index
    df = Repeat(df, over_bowled_index)
    
    ### More than 6 boundaries conceded filter
    boundaries_conceded = pd.DataFrame(df[df['total_runs'].isin([4, 5, 6, 7])].groupby(['match_id', 'bowler'])['total_runs'].value_counts())
    boundaries_conceded.rename(columns={'total_runs' : 'total_boundaries'}, inplace=True)
    boundaries_conceded.reset_index(inplace=True)
    boundaries_conceded.rename(columns={'total_runs' : 'boundary_type'}, inplace=True)
    boundaries_con_index = boundaries_conceded[(boundaries_conceded['boundary_type']==6) & (boundaries_conceded['total_boundaries']>2)].set_index(['match_id', 'bowler']).index
    df = Repeat(df, boundaries_con_index)
    
    ### More than 2 6s conceded filter
    tot_boundaries = pd.DataFrame(boundaries_conceded.groupby(['match_id', 'bowler'])['total_boundaries'].sum())
    tot_boundaries_index = tot_boundaries[tot_boundaries['total_boundaries']>=6].index
    df = Repeat(df , tot_boundaries_index)
    
    economy = df.groupby(['match_id', 'bowler'])['total_runs'].sum() / df.groupby(['match_id', 'bowler'])['over'].unique().apply(lambda x : len(x))
    economy = pd.DataFrame(economy)
    economy.rename(columns={0 : 'economy'}, inplace=True)
    economy = economy[economy['economy']<9]
    return economy, df
```

```python
economy, df = Economy_bowlers(sub_data.copy())
```

```python
economy.groupby('bowler')['economy'].mean().plot.bar(title='Avg economy rate of all bowlers from both teams', figsize=(15, 5))
```

```python
economy1, df1 = Economy_bowlers(sub_data[sub_data['season']>2015])
```

```python
economy1.groupby('bowler')['economy'].mean().plot.bar(title='Avg economy rate of all bowlers from both teams after 2015', figsize=(15, 5))
```

```python
batsmanwise_economy = pd.DataFrame(df1.groupby(['bowler', 'batsman'])['total_runs'].sum()/df1.groupby(['bowler', 'batsman'])['ball'].count())
```

```python
batsmanwise_economy[batsmanwise_economy[0]>0].groupby(['bowler', 'batsman']).mean().sort_values(by = 0, ascending=False).tail(15)
```

### How many batsman in the match will score more than 30 runs??

```python
batsman_runs = pd.DataFrame(sub_data.groupby(['match_id', 'batsman'])['batsman_runs'].sum())
```

```python
batsman_index = batsman_runs[batsman_runs['batsman_runs']<=30].index
```

```python
sub_data3 = sub_data.copy()
```

```python
for i, j in batsman_index:
    sub_data3.drop(sub_data3[(sub_data3['match_id']==i) & (sub_data3['batsman']==j)].index, inplace=True)
```

```python
sub_data3.groupby(['match_id', 'batsman'])['batsman_runs'].sum().plot.bar(title='Runs scored by batsman above 30 in all matches', figsize=(15, 5))
```

```python
(sub_data3[sub_data3['season']>2015].groupby(['match_id', 'batsman'])['batsman_runs'].sum().groupby('batsman').count()/sub_data[sub_data['season']>2015].groupby(['match_id', 'batsman'])['batsman_runs'].sum().groupby('batsman').count()).fillna(0).sort_values(ascending=False).head(21).plot.bar(title='% Chances of scoring more than 30 runs')
```
