```python
dummy = pd.DataFrame(sub_data[sub_data['season']>2015].groupby(['match_id', 'batsman'])['ball'].count())
```

```python
index = dummy[dummy['ball']<=10].index
```

```python
sub_data2 = sub_data[sub_data['season']>2015]
```

```python
for i, j in index:
    sub_data2.drop(sub_data2[(sub_data2['match_id']==i) & (sub_data2['batsman']==j)].index, inplace=True)
```

```python
def Calculate(df):
    boundary_runs = df[df['batsman_runs'].isin([4, 5, 6, 7])].groupby(['match_id', 'batsman'])['batsman_runs'].sum()
    non_boundary_runs = df[df['batsman_runs'].isin([1, 2, 3])].groupby(['match_id', 'batsman'])['batsman_runs'].sum()
    total_runs = df.groupby(['match_id', 'batsman'])['batsman_runs'].sum()
    ratio = (boundary_runs - non_boundary_runs) / total_runs
    return ratio.sort_values(ascending=False)
```

```python
Calculate(sub_data2).plot.bar(title='Boundary/total runs ratio', figsize=(15,5))
```

```python
(((sub_data2.groupby(['match_id', 'batsman'])['batsman_runs'].sum()/sub_data2.groupby(['match_id', 'batsman'])['ball'].count())*100).sort_values(ascending=False)).plot.bar(title='Strike rate of batsmans', figsize=(15,5))
```
